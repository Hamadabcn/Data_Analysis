```python
import pandas as pd

df = pd.read_csv('Failures.csv')

print(df.info())
```

```python
import matplotlib.pyplot as plt

xpoints = df['Age']
ypoints = df['Failures']

plt.plot(xpoints, ypoints)

plt.xlabel('Age')
plt.ylabel('Failures')

plt.show()
```

```python
xpoints = df['Age']
y1points = df['Failures']
y2points = df['DaysOff']

plt.subplot(1, 2, 1)

plt.scatter(xpoints, y1points)

plt.xlabel('Age')
plt.ylabel('Failures')

plt.subplot(1, 2, 2)

plt.scatter(xpoints, y2points)

plt.xlabel('Age')
plt.ylabel('DaysOff')

plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df)
plt.show()
```

```python
df = pd.read_csv('StdMarks.csv')

dfcorrel = df[["Math", "Phy", "Lang"]]

correlations = dfcorrel.corr(method='pearson')

print(correlations)

fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)

cax = ax.matshow(correlations)

ticks = [0, 1, 2]

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(["Math", "Phy", "Lang"])
ax.set_yticklabels(["Math", "Phy", "Lang"])

plt.show()
```
