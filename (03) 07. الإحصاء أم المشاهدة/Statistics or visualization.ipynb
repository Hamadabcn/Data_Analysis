```python
import pandas as pd

df1 = pd.read_csv('data1.csv')

print(df1.describe())
```

```python
df2 = pd.read_csv('data2.csv')

print(df2.describe())
```

```python
df3 = pd.read_csv('data3.csv')

print(df3.describe())
```

```python
df4 = pd.read_csv('data4.csv')

print(df4.describe())
```

```python
skew1 = df1.skew()
print(skew1)

skew2 = df2.skew()
print(skew2)

skew3 = df3.skew()
print(skew3)

skew4 = df4.skew()
print(skew4)
```

```python
df1.plot(kind='density')

df2.plot(kind='density')

df3.plot(kind='density')

df4.plot(kind='density')
```

```python
import matplotlib.pyplot as plt

# choose the columns
xpoints1 = df1['x']
ypoints1 = df1['y']

xpoints2 = df2['x']
ypoints2 = df2['y']

xpoints3 = df3['x']
ypoints3 = df3['y']

xpoints4 = df4['x']
ypoints4 = df4['y']

plt.subplot(2, 2, 1)
plt.scatter(xpoints1, ypoints1)

plt.subplot(2, 2, 2)
plt.scatter(xpoints2, ypoints2)

plt.subplot(2, 2, 3)
plt.scatter(xpoints3, ypoints3)

plt.subplot(2, 2, 4)
plt.scatter(xpoints4, ypoints4)

plt.show()
```
