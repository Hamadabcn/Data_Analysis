```python
import pandas as pd

df = pd.read_csv('failures.csv')

print(df.info())
```

```python
import matplotlib.pyplot as plt

ypoints = df['Failures']

plt.plot(ypoints)

plt.show()
```

```python
plt.plot(ypoints, marker='o')
plt.show()
```

```python
plt.plot(ypoints, marker ='o', ls =':')
plt.show()
```

```python
plt.plot(ypoints, color = 'red', linewidth = '2', marker ='o', ls ='-.')
plt.show()
```

```python
y1 = df['Failures']
y2 = df['Age']
plt.plot(y1, marker ='o')
plt.plot(y2, marker ='o')
plt.show()
```

```python
df = pd.read_csv('Failures2.csv')
x = df['Machine']
y = df['Failures']

plt.xlabel('Machine')
plt.ylabel('Failures')

# draw the lines
plt.bar(x, y)
plt.show()
```

```python
plt.pie(y, labels = x)
plt.show()
```
